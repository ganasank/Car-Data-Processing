```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack
from scipy.signal import butter, lfilter, freqz, filtfilt
import pandas as pd
import os
import glob as glob
import xlsxwriter


Ref=pd.read_excel(r'numpyref.xlsx')
xls = pd.ExcelFile(r'Time.xlsx')
RelDur=np.zeros(len(Ref.TITLE_SHEET))
StrDur=np.zeros(len(Ref.TITLE_SHEET))
CalDur=np.zeros(len(Ref.TITLE_SHEET))
MatDur=np.zeros(len(Ref.TITLE_SHEET))

for i in range(0,len(Ref.TITLE_SHEET)):
    Time1 = pd.read_excel(xls, Ref.TITLE_SHEET[i],header=None)
    RelDur[i]=Time1[1][8]-Time1[1][7]
    StrDur[i]=Time1[1][12]-Time1[1][11]
    CalDur[i]=Time1[1][6]-Time1[1][5]
    MatDur[i]=Time1[1][10]-Time1[1][9]
    print("For participant ",Ref.TITLE_SHEET[i],"Rel Duration is ",RelDur[i])
    print("For participant ",Ref.TITLE_SHEET[i],"Str Duration is ",StrDur[i])
    print("For participant ",Ref.TITLE_SHEET[i],"Cal Duration is ",CalDur[i])
    print("For participant ",Ref.TITLE_SHEET[i],"Mat Duration is ",MatDur[i])
    
print("Minimum Rel Dur is ",np.amin(RelDur))
print("Minimum Str Dur is ",np.amin(StrDur))
print("Minimum Cal Dur is ",np.amin(CalDur))
print("Minimum Mat Dur is ",np.amin(MatDur))
```
